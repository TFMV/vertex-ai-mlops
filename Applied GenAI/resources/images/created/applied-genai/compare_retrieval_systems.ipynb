```python
databases = [d['database'] for d in metrics]
mins = [d['min'] for d in metrics]
maxs = [d['max'] for d in metrics]
means = [d['mean'] for d in metrics]
medians = [d['median'] for d in metrics]
std_devs = [d['std_dev'] for d in metrics]
p95s = [d['p95'] for d in metrics]
p99s = [d['p99'] for d in metrics]
```

```python
plt.figure(figsize=(12, 6))  # Adjust figure size as needed

plt.plot(databases, mins, marker='o', linestyle='-', color='blue', label='Min')
plt.plot(databases, maxs, marker='s', linestyle='-', color='red', label='Max')
plt.plot(databases, means, marker='^', linestyle='-', color='green', label='Mean')
plt.plot(databases, medians, marker='D', linestyle='-', color='purple', label='Median')
plt.plot(databases, std_devs, marker='*', linestyle='-', color='orange', label='Std Dev')
plt.plot(databases, p95s, marker='x', linestyle='-', color='brown', label='P95')
plt.plot(databases, p99s, marker='+', linestyle='-', color='gray', label='P99')

plt.xlabel('Database')
plt.ylabel('Seconds')
plt.title('Profile: Sequential Retrieval Tries With Test Environment')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Create figure and subplots (order reversed)
fig, (ax2, ax1) = plt.subplots(2, 1, sharex=True, figsize=(12, 8),
                               gridspec_kw={'height_ratios': [1, 1]})

# Plotting on the first subplot (ax1) - range 0 to 0.4
ax1.plot(databases, mins, marker='o', linestyle='-', color='blue', label='Min')
ax1.plot(databases, maxs, marker='s', linestyle='-', color='red', label='Max')
ax1.plot(databases, means, marker='^', linestyle='-', color='green', label='Mean')
ax1.plot(databases, medians, marker='D', linestyle='-', color='purple', label='Median')
ax1.plot(databases, std_devs, marker='*', linestyle='-', color='orange', label='Std Dev')
ax1.plot(databases, p95s, marker='x', linestyle='-', color='brown', label='P95')
ax1.plot(databases, p99s, marker='+', linestyle='-', color='gray', label='P99')

ax1.set_ylim(0, 0.4)  # Set y-axis limits for ax1
ax1.spines['top'].set_visible(False)  # Hide top spine

# Plotting on the second subplot (ax2) - range 0.4 to 10
ax2.plot(databases, mins, marker='o', linestyle='-', color='blue', label='Min')
ax2.plot(databases, maxs, marker='s', linestyle='-', color='red', label='Max')
ax2.plot(databases, means, marker='^', linestyle='-', color='green', label='Mean')
ax2.plot(databases, medians, marker='D', linestyle='-', color='purple', label='Median')
ax2.plot(databases, std_devs, marker='*', linestyle='-', color='orange', label='Std Dev')
ax2.plot(databases, p95s, marker='x', linestyle='-', color='brown', label='P95')
ax2.plot(databases, p99s, marker='+', linestyle='-', color='gray', label='P99')

ax2.set_ylim(0.4, 10)  # Set y-axis limits for ax2
ax2.spines['bottom'].set_visible(False)  # Hide bottom spine

# Break effect (switched ax1 and ax2)
d = .015  # how big to make the diagonal lines in axes coordinates
kwargs = dict(transform=ax2.transAxes, color='k', clip_on=False)
ax2.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal
ax2.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # top-right diagonal

kwargs.update(transform=ax1.transAxes)  # switch to the bottom axes
ax1.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal
ax1.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # bottom-right diagonal

# Other plot elements
ax1.set_xlabel('Database')
ax2.set_ylabel('Seconds (0.4-10)')
ax1.set_ylabel('Seconds (0-0.4)')
ax2.set_title('Profile: Sequential Retrieval Tries With Test Environment')
plt.xticks(rotation=45, ha='right')
ax2.legend(loc='upper left')  # Legend for ax2 (now the top plot)
ax1.legend(loc='upper right') # Legend for ax1 (now the bottom plot)
plt.tight_layout()
plt.show()
```
